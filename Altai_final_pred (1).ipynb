```python
def mean_ball(row):
    if row <= 5:
        if (row%1 !=0)&(row*1000%100 == 0):
            return row*10
        else:
            return row
    elif 5 < row <= 10:
        return row*10
    elif 100 < row <= 500:
        return row/10
    elif 500 < row <= 1000:
        return row/10 
    elif 1000 < row <= 2000:
        return row/10
    elif 2000 < row <= 3000:
        return row/100
    elif (3000 < row <= 5000)&(row%1000 != 0): # не захватил 5000 и 4000
        if row%100 != 0: 
            return row/1000
        else:
            return row/100
    elif row > 5000:
        return row/100
    else:
        return row
    
    
```

```python
def clear_text(df,column,row_dict):
        """Функция проводит поиск и по шаблонам ключей словаря и замену на его значения."""

        corpus = df[column]
        for key,val in row_dict.items():
            corpus = corpus.apply(lambda x: val if key in str(x) else x)
        corpus = corpus.fillna('Не указан')
            

        return corpus
```

```python
def type_inst(row):
    
    pat_school = re.compile(r'(Ш|ш)кол|СОШ|СШ| сш |ОШ|(Г|г)имназ|(О|о)ш|МБОУ|МОУ|СОПШ|ООУ|мназия|кадетский|'
                            '(С|с)редн|СОМШ|АКПЛ|ГБОУШ| УВК |СУНЦ|ГАОУ ЦО|Ага Хана|Малышево|УВК ШГ| СПН', flags=re.U)
    pat_inst = re.compile(r'((У|у)нив|(И|и)нститут|(А|а)кадемия|(К|к)онсерватория|ФГБОУ|ВПО|АлтГ|ВО| ГУ |'
                          'верситет|АГАУ|ВЗФЭИ|ВЗФИ|АГТУ|ГТУ| СГА|ТГУ|ГАГУ|Сеченова|БЮИ)', flags=re.U)
    pat_col = re.compile(r'((У|у)чилище|(К|к)олле|(Л|л)ицей|(Т|т)ехник|СОУ|ПУ|КГБОУ|СПО| спо |ПТУ| пу |'
                         'ПЛ-39|КГУ|КГБПОУ|СХТ|АТИВТ| МТТ| НКТ )', flags=re.U)
    if re.findall(pat_school, row):
        return 'среднее'
    elif re.findall(pat_inst, row):
        return 'высшее'
    elif re.findall(pat_col, row):
        return 'профессиональное'
    else:
        return row

        
```
